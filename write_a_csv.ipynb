```python
odds = [1, 3, 5, 7, 9]
stuff = []
for value in odds:
    stuff.append([value, value**2, value**3])
stuff
```

```python
with open('stuff.csv','w') as csvfile:
csv_writer=csv.writer(open('stuff.csv', 'w'))
```

```python
csv_writer.writerows(stuff)
help(csv_writer)
```
